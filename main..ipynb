```python
import pandas as pd
```

```python
df1 = pd.read_csv("./Form Responses/Response_1.csv")
df2 = pd.read_csv("./Form Responses/Response_2.csv")
df3 = pd.read_csv("./Form Responses/Response_3.csv")
df4 = pd.read_csv("./Form Responses/Response_4.csv")
df5 = pd.read_csv("./Form Responses/Response_5.csv")

all_df = [df1, df2, df3, df4, df5]
# list of df of the true dataset
df_true = []
# list of df for measure of agreement 
df_agree = []
```

```python
# Dropping the timestamp column 
for i in range(len(all_df)):
    all_df[i] = all_df[i].drop(columns=['Timestamp'])

# Dropping df1 first two rows
all_df[0] = all_df[0].iloc[2:].reset_index().drop(columns='index')
all_df[0].head()
```

```python
# get only iloc[2] for true dataset, and iloc[:2] for the agreement dataset
for i in range(len(all_df)):
    # true dataset 
    df_true.append(all_df[i].iloc[[2]])

    # agreement dataset
    df_agree.append(all_df[i].iloc[:2])
```

```python
# merge datasets
df_merge_agree = df_agree[0]
df_merge_true = df_true[0]
for i in range(1, len(all_df)):
    df_merge_agree = df_merge_agree.join(df_agree[i])
    df_merge_true = df_merge_true.join(df_true[i])
```

GETTING MEASURE OF AGREEMENT STARTS HERE

```python
# Get observed agreement
matches = df_merge_agree.iloc[0] == df_merge_agree.iloc[1]
no_agree = matches.sum() 

observed_agreement = no_agree/511
print(observed_agreement)
```

```python
# Getting count of each class
row_counts = df_merge_agree.apply(lambda row: row.value_counts(), axis=1)
print(row_counts)
```

```python
# Expected Agreement
late_0 = row_counts["Late"].iloc[0]/511
early_0 = row_counts["Early"].iloc[0]/511
intermediate_0 = row_counts["Intermediate"].iloc[0]/511

late_1 = row_counts["Late"].iloc[1]/511
early_1 = row_counts["Early"].iloc[1]/511
intermediate_1 = row_counts["Intermediate"].iloc[1]/511

expected_agreement = (late_0 * late_1) + (early_0 * early_1) + (intermediate_0 * intermediate_1)
print("Expected Agreement: ",expected_agreement)
```

```python
# Calculating Cohen Kappa's Agreement 
kappa = (observed_agreement - expected_agreement)/(1-expected_agreement)
if kappa >= 0.81:
    print("Almost perfect agreement: ", kappa)
elif kappa >= 0.61:
    print("Substantial agreement: ", kappa)
elif kappa >= 0.41:
    print("Moderate agreement: ", kappa)
elif kappa >= 0.21:
    print("Fair agreement: ", kappa)
elif kappa >= 0:
    print("Slight agreement: ", kappa)
else:
    print("Poor agreement: ", kappa)
```
