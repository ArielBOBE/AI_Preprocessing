```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading csv
df = pd.read_csv('./Form Responses/wide_resolved_data.csv')
df.head()
```

```python
long_df = pd.melt(df, value_name = "Severity")
long_df.rename(columns={"variable":"Image"})
long_df.head()
```

```python
# Dropping everything that isn't late, early, or intermediate
long_df = long_df[long_df["Severity"].isin(["Early", "Intermediate", "Late"])]
```

```python
class_distribution = long_df['Severity'].value_counts()
print(class_distribution)
```

```python
# Plotting the distribution graph
categories = ["Early", "Intermediate", "Late"]
values = [class_distribution.iloc[1], class_distribution.iloc[2], class_distribution.iloc[0]]

plt.bar(categories, values, color="blue", ec="black")
plt.title("Severity Class Distribution")
plt.xlabel("Severity")
plt.ylabel("Frequency")
plt.show()
```
